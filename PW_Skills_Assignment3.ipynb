1. Reverse a string:

2. Count the number of vowels in a string:
   ```python
   def count_vowels(s):
       vowels = "aeiouAEIOU"
       return sum(1 for char in s if char in vowels)
   ```

3. Check if a given string is a palindrome:

5. Find all occurrences of a substring:
   ```python
   def find_substring_occurrences(s, sub):
       return [i for i in range(len(s) - len(sub) + 1) if s[i:i+len(sub)] == sub]
   ```

6. Basic string compression using counts of repeated characters:
   ```python
   def compress_string(s):
       compressed = []
       count = 1
       for i in range(1, len(s)):
           if s[i] == s[i - 1]:
               count += 1
           else:
               compressed.append(s[i - 1] + str(count))
               count = 1
       compressed.append(s[-1] + str(count))
       return ''.join(compressed)
   ```

12. Find the second largest number in a list:
    ```python
    def second_largest(lst):
    ```

14. Reverse a list in-place without using built-in reverse functions:
    ```python
    def reverse_list(lst):
        for i in range(len(lst) // 2):
            lst[i], lst[-i - 1] = lst[-i - 1], lst[i]
        return lst
    ```

15. Find and remove duplicates from a list while preserving order:
    ```python
    def remove_duplicates(lst):
        seen = set()
        return [x for x in lst if not (x in seen or seen.add(x))]
    ```

20. Shuffle a list randomly without using built-in shuffle functions:
    ```python
    import random

    def custom_shuffle(lst):
        shuffled = lst[:]
        for i in range(len(shuffled) - 1, 0, -1):
            j = random.randint(0, i)
            shuffled[i], shuffled[j] = shuffled[j], shuffled[i]
        return shuffled
    ```

32. Merge two dictionaries, adding values for common keys:
    ```python
    def merge_dictionaries(dict1, dict2):
        merged = dict1.copy()
        for key, value in dict2.items():
            merged[key] = merged.get(key, 0) + value
        return merged
    ```

33. Access a value in a nested dictionary with a list of keys:
    ```python
    def access_nested_dict(d, keys):
        for key in keys:
            if isinstance(d, dict) and key in d:
                d = d[key]
            else:
                return None
        return d
    ```

34. Return a sorted version of a dictionary by values:
    ```python
    def sort_dict_by_values(d, ascending=True):
        return dict(sorted(d.items(), key=lambda item: item[1], reverse=not ascending))
    ```

35. Invert a dictionary, handling duplicate values:
    ```python
    def invert_dictionary(d):
        inverted = {}
        for key, value in d.items():
            if value in inverted:
                inverted[value].append(key)
            else:
                inverted[value] = [key]
        return inverted
    ```
