QUE 1. Explain the key features of Python that make it a popular choice for programming
Python is one of the most popular programming languages because of its versatility, ease of use, and readability. Some of the key features that make Python a popular choice are:

2. Interpreted Language: Python code is executed line by line, making debugging easier.
3. Object-Oriented: Python supports object-oriented programming, allowing developers to create and manage objects efficiently.

QUE 2. Describe the role of predefined keywords in Python and provide examples of how they are used in a program.
Predefined keywords in Python are reserved words that have specific meanings and purposes. These keywords cannot be used as variable names or identifiers.

Examples of Keywords:
- if, else, elif: Used for conditional statements.
- for, while: Used for looping.
- def: Used to define a function.
- return: Used to return a value from a function.
- class: Used to define a class.

Example:
```python
def check_number(n):
    if n > 0:
        return "Positive"
    elif n < 0:
        return "Negative"
    else:
        return "Zero"

print(check_number(5))  # Output: Positive
```

In this program, keywords like `def`, `if`, `elif`, and `else` are used to define the function and control flow.




QUE 3. Compare and contrast mutable and immutable objects in Python with examples
Mutable vs Immutable Objects in Python
In Python, objects are classified into two types based on whether their content can be changed:

1. Mutable Objects: These are objects whose value can be changed after they are created.
   - Examples: Lists, dictionaries, sets.
   - Example:
   ```python
   my_list = [1, 2, 3]
   my_list.append(4)
   print(my_list)  # Output: [1, 2, 3, 4]
   ```

2. Immutable Objects: These are objects whose value cannot be changed once they are created.
   - Examples: Integers, strings, tuples.
   - Example:
   ```python
   my_tuple = (1, 2, 3)
   # my_tuple[0] = 4  # This would raise an error as tuples are immutable.
   ```

Comparison:
- Mutable objects can be altered, while immutable objects cannot.
- Mutable objects are often used when you need to modify data, while immutable objects provide a layer of safety by preventing accidental changes.


QUE 4. Explain the concept of type casting in Python with examples
Type casting in Python refers to converting one data type into another. This can be useful when you need to perform operations on incompatible data types. Python provides built-in functions to perform type casting.

Types of Type Casting:
Explicit Type Casting: Done using functions like int(), float(), str(), list(), etc.
Implicit Type Casting: Python automatically converts one data type to another without the need for explicit instructions.

Examples:
1. int(): Converts a value to an integer.
2. float(): Converts a value to a floating-point number.
3. str(): Converts a value to a string.
4. list(): Converts a value to a list.


QUE 5. Discuss the different types of operators in Python and provide examples of how they are used
Python provides various operators for different operations like arithmetic, comparison, logical, etc.

Types of Operators:

1. Arithmetic Operators: Perform mathematical operations.
   - `+`, `-`, `*`, `/`, `%`, `**` (exponentiation), `//` (floor division)
   - Example:
     ```python
     x = 10
     y = 3
     print(x + y)  # Output: 13
     print(x ** y)  # Output: 1000 (10 raised to the power of 3)
     ```

2. Comparison (Relational) Operators: Compare two values and return `True` or `False`.
   - `==`, `!=`, `>`, `<`, `>=`, `<=`
   - Example:
     ```python
     a = 5
     b = 10
     print(a > b)  # Output: False
     print(a != b)  # Output: True
     ```

3. Logical Operators: Combine conditional statements.
   - `and`, `or`, `not`
   - Example:
     ```python
     age = 25
     print(age > 18 and age < 30)  # Output: True
     ```

4. Assignment Operators: Assign values to variables.
   - `=`, `+=`, `-=`, `*=`, `/=`
   - Example:
     ```python
     x = 10
     x += 5  # Equivalent to x = x + 5
     print(x)  # Output: 15
     ```

5. Bitwise Operators: Perform bit-level operations.
   - `&`, `|`, `^`, `~`, `<<`, `>>`
   - Example:
     ```python
     x = 6  # 110 in binary
     y = 3  # 011 in binary
     print(x & y)  # Output: 2 (010 in binary)




QUE 6. How do conditional statements work in Python? Illustrate with examples
Conditional statements allow you to control the flow of execution based on certain conditions. Python uses `if`, `elif`, and `else` for condition-based logic.

Example 1: Basic `if-else` Statement
```python
age = 18
if age >= 18:
    print("Adult")
else:
    print("Minor")
```
If the `age` is greater than or equal to 18, it prints "Adult"; otherwise, it prints "Minor."

Example 2: `if-elif-else` Statement
```python
marks = 85
if marks >= 90:
    print("Grade A")
elif marks >= 75:
    print("Grade B")
else:
    print("Grade C")
```
The program checks multiple conditions using `if-elif-else` and returns the appropriate grade based on the value of `marks`.

Example 3: Nested Conditional Statement
```python
x = 10
if x > 0:
    if x % 2 == 0:
        print("Positive Even Number")
    else:
        print("Positive Odd Number")
else:
    print("Negative Number")
```
Here, nested conditions are used to check if the number is positive and whether it is even or odd.


QUE 7. Describe the different types of loops in Python and their use cases with examples
Loops are used to execute a block of code repeatedly until a certain condition is met. Python provides two types of loops:

1. For Loop: Used to iterate over a sequence (like a list, tuple, or string).

Example 1: Basic `for` loop
```python
for i in range(5):
    print(i)
# Output: 0 1 2 3 4
```

Example 2: Iterating over a list
```python
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(fruit)
# Output: apple, banana, cherry
```

Use Cases:
When you know the number of iterations in advance (like looping through elements in a list or range).

2. While Loop: Repeats a block of code as long as the condition is `True`.

Example 1: Basic `while` loop
```python
count = 0
while count < 5:
    print(count)
    count += 1
# Output: 0 1 2 3 4
```

Example 2: `while` loop with user input
```python
password = ""
while password != "secret":
    password = input("Enter password: ")
print("Access granted")
```

Use Cases:
When the number of iterations is not predetermined (e.g., waiting for user input or processing until a condition becomes `False`).

3. Nested Loops: Loops inside another loop.
```python
for i in range(3):
    for j in range(2):
        print(f"i: {i}, j: {j}")
```
