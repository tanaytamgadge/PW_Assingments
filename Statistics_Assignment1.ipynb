1. Types of Data:
Qualitative Data: Descriptive information that cannot be measured numerically.


Examples: Colors (red, blue), Types of cuisine (Italian, Mexican).
Scales:
Nominal Scale: Categories without a specific order. Example: Gender (male, female).
Ordinal Scale: Ordered categories. Example: Customer satisfaction (poor, good, excellent).
Quantitative Data: Numerical information that can be measured.


Examples: Height (cm), Salary ($).
Scales:
Interval Scale: Equal intervals without a true zero. Example: Temperature in Celsius.
Ratio Scale: Equal intervals with a true zero. Example: Weight (kg).

2. Measures of Central Tendency:
Mean: Average of data.
Example: Test scores (50, 60, 70 → Mean = 60).
Use: When data is symmetrical and without extreme outliers.
Median: Middle value of data.
Example: Test scores (50, 60, 100 → Median = 60).
Use: When data is skewed or contains outliers.
Mode: Most frequent value.
Example: Test scores (50, 50, 70 → Mode = 50).
Use: For categorical data or when identifying the most common value is required.

3. Concept of Dispersion:
Dispersion measures the spread of data.
Variance: Average of squared differences from the mean.
Formula: σ2=∑(xi−μ)2N\sigma^2 = \frac{\sum (x_i - \mu)^2}{N}.
Standard Deviation: Square root of variance, provides spread in the same units as data.
Higher values indicate greater spread.

4. Box Plot:
A graphical representation of data distribution.
Components:
Minimum, Q1 (First Quartile), Median, Q3 (Third Quartile), Maximum.
Outliers: Points outside 1.5 times the IQR.
Insights:
Spread, skewness, and presence of outliers.

5. Role of Random Sampling:
Ensures every member of a population has an equal chance of selection.
Importance:
Reduces bias.
Ensures a representative sample for generalization to the population.

6. Skewness:
Describes the asymmetry of data distribution.
Types:
Positive Skew: Tail on the right (e.g., income data).
Negative Skew: Tail on the left (e.g., exam scores with many high performers).
Affects the mean and median, impacting summary statistics.

7. Interquartile Range (IQR):
Measures the spread of the middle 50% of data.
Formula: IQR=Q3−Q1\text{IQR} = Q3 - Q1.
Use:
Detects outliers: Data points outside


8. Binomial Distribution:
Used when:
Two possible outcomes (success/failure).
Fixed number of trials.
Independent trials with constant probability.
Example: Flipping a coin 10 times.

9. Normal Distribution and Empirical Rule:
Properties:
Symmetrical, bell-shaped curve.
Mean = Median = Mode.
Empirical Rule:
68% within 1 SD, 95% within 2 SD, 99.7% within 3 SD of the mean.

10. Poisson Process Example:
Real-life Example: Number of cars passing a toll booth in an hour.
Average rate (λ\lambda) = 10 cars/hour.
Probability of exactly 5 cars:
Formula: P(X=k)=e−λλkk!P(X=k) = \frac{e^{-\lambda} \lambda^k}{k!}.
P(5)=e−10⋅1055!≈0.037P(5) = \frac{e^{-10} \cdot 10^5}{5!} \approx 0.037.

11. Random Variables:
Definition: Variable representing outcomes of a random process.
Discrete: Countable outcomes.
Example: Number of heads in 10 coin flips.
Continuous: Infinite possible values.
Example: Time taken to complete a task.



 12. Provide an example dataset, calculate both covariance and correlation, and interpret the results
 Example Dataset:
Study Hours (X)       Test Scores (Y)
      2                      50
      4                      55
      6                      60
      8                      65
      10                     70


Covariance:
Formula:
 Covariance = Sum of [(X - Mean of X) × (Y - Mean of Y)] ÷ Number of observations
Calculate the means:
Mean of X is 6.
Mean of Y is 60.
Compute deviations and products:
Deviations from means are calculated for each value of X and Y.
Multiply the deviations for each pair, then sum them (result = 100).
Divide by the number of values (5):
Covariance = 100 ÷ 5 = 20.

Correlation:
Formula:
Correlation = Covariance ÷ (Standard deviation of X × Standard deviation of Y)
Standard deviations:
Standard deviation of X is 2.83.
Standard deviation of Y is 7.07.
Calculate correlation:
Correlation = 20 ÷ (2.83 × 7.07) = 1.

Interpretation:
Covariance: Positive value (20) indicates that as study hours increase, test scores also increase.
Correlation: A value of 1 means there is a perfect positive linear relationship between study hours and test scores.
